```python
import matplotlib.pyplot as plt
import numpy as np
#plt.style.use("ggplot")
```

## Atractor de Lorentz

```python
x0 = 1
y0 = 1
z0 = 1
a = 10
b = 16
c = 1
dt = 0.001
i = 50000

xn = [x0]
yn = [y0]
zn = [z0]

for n in range(i):
    xn.append(x0 + dt*a*(y0 - x0))
    yn.append(y0 + dt*(b*x0 - y0 -x0*z0))
    zn.append(z0 + dt*(x0*y0 - c*z0))
    x0 = xn[n+1]
    y0 = yn[n+1]
    z0 = zn[n+1]
plt.plot(xn, zn, 'k.', markersize=0.5)
```

## Atractor de Pickover

```python
x0 = -3
y0 = -2
h = -0.02
i = 100000

xn = [x0]
yn = [y0]

for n in range(i):
    xn.append(x0 - h*np.sin(y0 + np.tan(3*y0)))
    yn.append(y0 - h*np.sin(x0 + np.tan(3*x0)))
    x0 = xn[n+1]
    y0 = yn[n+1]
plt.plot(xn, yn, 'k.', markersize=0.5)
```

## Atractor de Hnon

```python
x0 = -0.01
y0 = -0.01
a = 1.4
b = 0.3
i = 10000

xn = [x0]
yn = [y0]

for n in range(i):
    xn.append(1 - a*x0*x0 + y0)
    yn.append(b*x0)
    x0 = xn[n+1]
    y0 = yn[n+1]
plt.plot(xn, yn, 'k.', markersize=0.5)
```

## Atractor con forma de mueco

```python
x0 = -0.01
y0 = -0.01
i = 10000

xn = [x0]
yn = [y0]

for n in range(i):
    xn.append(1 - y0 + abs(x0))
    yn.append(x0)
    x0 = xn[n+1]
    y0 = yn[n+1]
plt.plot(xn, yn, 'k.', markersize=0.5)
```
